# Matplotlib

Matplotlib ist eine Bibliothek zum Plotten. Wir verwenden das Untermodul `pyplot`. pyplot stellt eine prozedurale Schnittstelle zur Verf√ºgung. Ein Graph in Matplotlib ist eine zwei- oder dreidimensionale Zeichnung, die mit Hilfe von Punkten, Kurven, Balken oder anderem einen Zusammenhang herstellt. 

Doku: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot

```python
import matplotlib.pyplot as plt
plt.plot([-1, -4.5, 16, 23, 15, 59])
plt.show()
```

## Formatparameter

Indem wir einen Formatstring beim Funktionsaufruf mit√ºbergeben, k√∂nnen wir einen Graphen mit diskreten Werten erzeugen. Der Formatparameter hat das folgende Format: `[marker][line][color]` F√ºr `marker` werden u.a. folgende Werte angenommen `.`, `,`, `o`, `v`, f√ºr `line` u.a. `-`, `--`, `-.`, `:` und f√ºr `color` u.a. `r`, `g`, `b`, `c`, `m`, `y`, `k`, `w`. Somit w√§re `or` ein valider Formatparameter.

```python
x = [1, 3, 4, 6,  7]
y = [3, 6, 8, 10, 13]
plt.plot(x,y, 'or')
plt.title("Das ist der Titel")
plt.xlabel('x-Achse')
plt.ylabel('y-Achse')
plt.show()
```

## Scatter

Mit der `scatter` lassen sich ebenfalls diskrete Werte abbilden. Scatter erlaubt jedoch genauere Visualisierung von eizelnen Punkten. 

```python
x1 = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y1 = [99,86,87,88,111,86,103,87,94,78,77,85,86]
sizes = [20,50,100,200,500,1000,60,90,10,300,600,800,75]
colors = ["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"]

plt.scatter(x1,y1,c=colors,s=sizes)
plt.show()
```
